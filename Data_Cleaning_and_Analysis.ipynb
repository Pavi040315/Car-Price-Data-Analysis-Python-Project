# Task 3:  Data Cleaning & Data Processing

```python
# Import pandas library

import pandas as pd
```

```python
# Load and display the car_prices dataset

df = pd.read_csv("car_prices.csv")
df
```

```python
# Return uniques from the column 'seller' 

df.seller.unique() 
```

```python
# Return number of uniques from the column 'seller' before cleaning the column

df.seller.nunique() 
```

```python
# Replace '-' with ' '(whitespace) e.g.'nissan-infiniti lt' replace with 'nissan-infiniti lt'

df.seller = df.seller.str.replace("-", " ").str.replace("  ", " ")  
df
```

```python
# Return number of uniques from the column 'seller' after cleaning the column

df.seller.nunique() 
```

```python
# Discover duplicates

df.duplicated()
```

```python
# Discover number of duplicates

df.duplicated().sum()
```

No duplicates have been found

```python
# Discover empty cells

df.isnull()
```

```python
# Discover number of empty cells

df.isnull().sum()
```

```python
# Remove rows that contain empty cells and write in the same DataFrame

df.dropna(inplace = True)
df
```

```python
# Check whether have any empty cells

df.isnull().sum()
```

Dataset has been cleaned as the empty cells have been dropped and no duplicates found.

## Task 5: Create the visuals using Matplotlib or Bokeh

### a. Choose the suitable type of chart to plot an interactive chart for the number of cars sold over year after 2000.

```python
# Interactive Horizontal Bar Chart
import pandas as pd
import matplotlib.pyplot as plt

# Interactive visualization
%matplotlib notebook
df = pd.read_csv("car_prices.csv")
all_year = df.groupby('year').size().reset_index(name = 'number_of_cars_sold')
year_greater_2000 = all_year[all_year['year'] > 2000].reset_index()

fig = plt.subplots(figsize =(10, 5)) # Change the dimension of the plot
x = year_greater_2000.year
y = year_greater_2000.number_of_cars_sold
plt.title("Total Number Of Cars Sold Over Year After 2000") # Title
plt.xlabel("Number of cars sold") # x-axis label
plt.ylabel("Year") # y-axis label
plt.xticks(range(0,120000,10000))
plt.yticks(range(2001,2016,1))
plt.barh(x,y, color='greenyellow') # color change
plt.show()
```

### b. Choose the suitable type of chart to plot an interactive chart for the top 10 seller to their total selling price.

```python
#Bar chart & Scatterplot - HoverTool
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool

# Read data from data set
df = pd.read_csv("car_prices.csv")
df.seller = df.seller.str.replace("-", " ").str.replace("  ", " ")
grouped_seller = df.groupby('seller')['sellingprice'].sum().reset_index().sort_values(by='sellingprice', ascending=False).head(10)

x=grouped_seller.seller
y=grouped_seller.sellingprice


p=figure(x_range=x, width=900, height=600, toolbar_location='right')
# plot a bar chart
p.vbar(x, top= y,width=0.3, color='limegreen')
#p.scatter(x= x, y= y)

p.title.text="Total Number Of Cars Sold Over Year After 2000"
p.title.align="center"
p.xaxis.axis_label = "Seller"
p.yaxis.axis_label = "Selling Price"

p.xaxis.major_label_orientation = 45
p.scatter(x= x, y= y,fill_color="white", size=10)

# Create a hover tool called hover
hover = HoverTool(tooltips=[ ('Seller', '@x'), ('Selling Price','@y')])

# Add the hover tool to the plot
p.add_tools(hover)

# show the results
output_notebook()
show(p)
```

#### (ii) Total Number Of Cars Sold According to Their Transmission

```python
#Bar chart & Scatterplot - HoverTool
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool

# Read data from data set
df = pd.read_csv("car_prices.csv")
prefer_transmission = df.transmission.value_counts().reset_index(name = 'Total_number_of_cars_sold')
prefer_transmission = prefer_transmission[(prefer_transmission['transmission'] != 'sedan') & (prefer['transmission'] != 'Sedan')]

# Set x axis and y axis variables
x=prefer_transmission.transmission
y=prefer_transmission.Total_number_of_cars_sold

name = 'number_of_cars_sold'
p=figure(x_range=x, width=800, height=500, toolbar_location='right')
# plot a bar chart
p.vbar(x, top= y,width=0.3, color='red')
#p.scatter(x= x, y= y)

p.title.text="Total Number Of Cars Sold According to Their Transmission"
p.title.align="center"
p.xaxis.axis_label = "Transmission"
p.yaxis.axis_label = "Cars Sold"

p.xaxis.major_label_orientation = 45
plt.yticks(range(0,550000,10000))
p.scatter(x= x, y= y,fill_color="white", size=10)

# Create a hover tool called hover
hover = HoverTool(tooltips=[ ('Transmission', '@x'), ('Cars Sold','@y')])

# Add the hover tool to the plot
p.add_tools(hover)

# show the results
output_notebook()
show(p)
```
